```python
import pandas as pd
import numpy as np
df = pd.read_csv('/content/emot_train_cleaned_orig_backtranslated_en_cleaned.csv')
print(df)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize

def assemble_documents(df, text_column):

  documents = []
  for _, row in df.iterrows():
    text = row[text_column]
    sentences = sent_tokenize(text)
    documents.append(" ".join(sentences))
  return documents

documents = assemble_documents(df, 'tweet')

print("Number of documents assembled:", len(documents))
```

```python
from nltk.tokenize import sent_tokenize
nltk.download('punkt')

tokenized_documents = []
for document in documents:
    tokens = sent_tokenize(document)
    tokenized_documents.append(tokens)

for tokens in tokenized_documents:
    print(tokens)
    print("\n")
```

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')

tokenized_documents = []
for document in documents:
    tokens = word_tokenize(document)
    tokenized_documents.append(tokens)

for tokens in tokenized_documents:
    print(tokens)
    print("\n")
```

```python
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

all_tokens = [token for document_tokens in tokenized_documents for token in document_tokens]

fdist = FreqDist(all_tokens)

print(fdist.most_common(10))

fdist.plot(30,cumulative=False)
plt.show()
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_tokenized_documents = []
for document_tokens in tokenized_documents:
  filtered_tokens = [token for token in document_tokens if token.lower() not in stop_words]
  filtered_tokenized_documents.append(filtered_tokens)

for tokens in filtered_tokenized_documents:
  print(tokens)
  print("\n")
```

```python
import string

def remove_punctuation(text):
  """Removes punctuation from a string."""
  return text.translate(str.maketrans('', '', string.punctuation))

df['tweet_no_punctuation'] = df['tweet'].apply(remove_punctuation)

print(df[['tweet', 'tweet_no_punctuation']])
```

```python
df['text_uppercase'] = df['tweet'].str.upper()

df['text_lowercase'] = df['tweet'].str.lower()

print(df)
```

```python
!pip install -U spacy
!python -m spacy download en_core_web_sm

import spacy
nlp = spacy.load("en_core_web_sm")

def annotate_text(text):
  doc = nlp(text)
  annotations = []
  for token in doc:
    annotations.append({
        'text': token.text,
        'lemma': token.lemma_,
        'pos': token.pos_,
        'tag': token.tag_,
        'dep': token.dep_,
        'shape': token.shape_,
        'is_alpha': token.is_alpha,
        'is_stop': token.is_stop
    })
  return annotations


df['annotations'] = df['tweet'].apply(annotate_text)

print(df[['tweet', 'annotations']])

```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()


def stem_tokens(tokens):
  stemmed_tokens = [stemmer.stem(token) for token in tokens]
  return stemmed_tokens


def lemmatize_tokens(tokens):
  lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
  return lemmatized_tokens


df['stemmed_tokens'] = df['tweet'].apply(lambda text: stem_tokens(word_tokenize(text)))
df['lemmatized_tokens'] = df['tweet'].apply(lambda text: lemmatize_tokens(word_tokenize(text)))

print(df[['tweet', 'stemmed_tokens', 'lemmatized_tokens']])
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')

def pos_tag_text(text):
  tokens = nltk.word_tokenize(text)
  tagged_tokens = nltk.pos_tag(tokens)
  return tagged_tokens

df['pos_tags'] = df['tweet'].apply(pos_tag_text)

print(df[['tweet', 'pos_tags']])
```

```python
def ner_text(text):
  doc = nlp(text)
  entities = [(ent.text, ent.label_) for ent in doc.ents]
  return entities

df['ner_entities'] = df['tweet'].apply(ner_text)

print(df[['tweet', 'ner_entities']])
```

```python
from nltk import ngrams

def generate_ngrams(text, n):
  tokens = word_tokenize(text)
  n_grams = list(ngrams(tokens, n))
  return n_grams

df['unigrams'] = df['tweet'].apply(lambda text: generate_ngrams(text, 1))
df['bigrams'] = df['tweet'].apply(lambda text: generate_ngrams(text, 2))
df['trigrams'] = df['tweet'].apply(lambda text: generate_ngrams(text, 3))

print(df[['tweet', 'unigrams', 'bigrams', 'trigrams']])
```
