```python
numeric_cols = df.select_dtypes(include=np.number).columns
z_scores = np.abs((df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std())
threshold = 3
outliers = (z_scores > threshold).any(axis=1)
print(df[outliers])

import matplotlib.pyplot as plt
from scipy import stats
plt.figure(figsize=(10, 6))
for col in numeric_cols:
  plt.scatter(df.index[~outliers], df.loc[~outliers, col])
  plt.scatter(df.index[outliers], df.loc[outliers, col], label='Outliers', color='red')
  plt.xlabel('Index')
  plt.ylabel(col)
  plt.title(f'Outlier Detection for {col}')
  plt.legend()
  plt.show()
```

```python
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1
outlier_condition = (df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))
outliers_iqr = outlier_condition.any(axis=1)
print(df[outliers_iqr])

plt.figure(figsize=(10, 6))
for col in numeric_cols:
    hist_data, bin_edges = np.histogram(df.loc[outliers, col])
    plt.hist(bin_edges[:-1], bin_edges, weights=hist_data, label='Outliers', color='red')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f'Outlier Detection for {col} (IQR)')
    plt.legend()
    plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.covariance import EllipticEnvelope

numeric_cols = d.select_dtypes(include=np.number).columns
data = d[numeric_cols]
model = EllipticEnvelope(contamination=0.05)
model.fit(data)
predictions = model.predict(data)
outliers_mahalanobis = predictions == -1
print(d[outliers_mahalanobis])

plt.figure(figsize=(10, 6))
for col in numeric_cols:
    plt.boxplot([data.loc[~outliers_mahalanobis, col], data.loc[outliers_mahalanobis, col]],
                labels=['Inliers', 'Outliers'], patch_artist=True,
                boxprops=dict(facecolor='lightblue'),
                whiskerprops=dict(color='blue'),
                capprops=dict(color='blue'),
                medianprops=dict(color='red'),
                flierprops=dict(marker='o', markerfacecolor='red', markersize=8))
    plt.xlabel('Index')
    plt.ylabel(col)
    plt.title(f'Outlier Detection using Mahalanobis Distance for {col}')
    plt.legend()
    plt.show()
```

```python
from scipy import stats

d = pd.read_csv('/content/population (1).csv')
df = d.reset_index()
df['Z-Score'] =np.abs(stats.zscore(df['Value']))
threshold = 3
df['outlier'] = df['Z-Score'] > threshold
print(df['outlier'])

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Value'], label='Year', color='blue')
plt.scatter(df.index[df['outlier']], df['Value'][df['outlier']], color='red', label='Outliers')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Time series with outlier')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, chi2

X = d.drop('Value', axis=1)
y = d['Value']
numeric_cols = X.select_dtypes(include=['float', 'int']).columns
k = min(3, len(numeric_cols))
k_best = SelectKBest(score_func=chi2, k=k)
X_new = k_best.fit_transform(X[numeric_cols], y)
print(X_new)
selected_feature_indices = k_best.get_support(indices=True)
selected_features = numeric_cols[selected_feature_indices]

plt.figure(figsize=(10, 6))
for feature in selected_features:

  plt.hist(d[feature], bins=10)
  plt.xlabel(feature)
  plt.ylabel('Frequency')
  plt.title(f'Histogram of {feature}')
  plt.show()
```
