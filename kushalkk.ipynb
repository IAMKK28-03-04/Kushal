```python
#USE MAP FUNCTION TO PANDAS DATA FRAME

    # Map a function to a specific column
df['Value'] = df['Value'].map(lambda x: x * 2)

    # Do something with the row data
df['New Value'] = df.apply(my_function, axis=1)

    #Map a function to a specific column with conditions
df['Value'] = df['Value'].map(lambda x: x * 2 if x > 1000000 else x)

    # Map a function to multiple columns
df['New Value'] = df.apply(lambda row: row['Value'] + row['Value'], axis=1)
print(df)
```

```python
    # Map a function to the 'Value' column

def get_decade(x):
    return x - (x % 10)
df['Decade'] = df['Year'].map(get_decade)
print(df)
```

```python
    # Map a function to the 'Value' column

def convert_to_millions(x):
    return x / 1_000_000
df['Value'] = df['Value'].map(convert_to_millions)
print(df)
```

```python
# USE FILTER FUNCTION TO PANDAS DATA FRAME

    #Filter by country
aruba_df = df[df['Country Name'] == 'Aruba']

    #Filter by value range
large_df = df[df['Value'] > 1000000]

    #Filter by year range
recent_df = df[(df['Year'] >= 2000) & (df['Year'] <= 2020)]

    #Filter by country and year
zimbabwe_2019_df = df[(df['Country Name'] == 'Zimbabwe') & (df['Year'] == 2019)]
print(df)
```

```python
   #Filter based on Country Name

df = df[df['Country Name'] == 'Zimbabwe']
print(df)
```

```python
    #Filter by Multiple Conditions

df = df.loc[((df['Country Name'] == 'Zimbabwe') | (df['Country Name'] == 'Aruba')) & (df['Year'] == 2020)]
print(df)
```

```python
# USE REDUCE FUNCTION TO PANDAS DATA FRAME

   #Calculate the total population
total_population = df['Value'].sum()

    #Calculate the maximum population for each country
max_population = df.groupby('Country Name')['Value'].max()

   #Calculate the average population per country
average_population = df.groupby('Country Name')['Value'].mean()

   #Calculate the total population for each year
yearly_population = df.groupby('Year')['Value'].sum()
print(df)
```

```python
# USE LAMBDA FUNCTION TO PANDAS DATA FRAME

     #Convert the 'Value' column to millions
df['Value'] = df['Value'].apply(lambda x: x / 1_000_000)

     #Calculate the population growth rat
df['Growth Rate'] = df.apply(lambda row: (row['Value'] - df.loc[row.name-1, 'Value']) / df.loc[row.name-1, 'Value'] if row.name > 0 else 0, axis=1)

     # Get the country name with the highest population
max_population_country = max(df.apply(lambda row: row['Country Name'] if row['Value'] == df['Value'].max() else '', axis=1))

     # Get the year with the highest population
max_population_year = max(df.apply(lambda row: row['Year'] if row['Value'] == df['Value'].max() else '', axis=1))
print(df)
```

```python
    # Using apply with a lambda function to process multiple columns

df['Country Year'] = df.apply(lambda row: f"{row['Country Name']} {row['Year']}", axis=1)
print(df)
```

```python
    # Update 'Country Code' to lowercase

df['Country Code'] = df['Country Code'].map(lambda x: x.lower())
print(df)
```

```python
# VISUALIZE THE DATA SET (At least 6 different plots)
               # World Population Dataset
#Scatter Plot

import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('/content/population.csv')
country_data = data[data['Country Name'] == 'Aruba']
plt.figure(figsize=(10, 6))
plt.plot(country_data['Year'], country_data['Value'], marker='o')
plt.title('Population Value Over Years for Aruba')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.show()

```

```python
   #Bar Plot

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/population.csv')
year_data = data[data['Year'] == 2021]
plt.figure(figsize=(12, 8))
sns.barplot(x='Country Name', y='Value', data=year_data)
plt.title('Population Value by Country in 2021')
plt.xlabel('Country')
plt.ylabel('Value')
plt.xticks(rotation=90)
plt.show()
```

```python
    #Histogram

import matplotlib.pyplot as plt
for country in data['Country Name'].unique():
    country_data = data[data['Country Name'] == country]
    plt.hist(country_data['Value'], bins=50, alpha=0.5, label=country)
plt.title('Histogram of Values by Country')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
   #Box Plot
plt.figure(figsize=(12, 6))
plt.boxplot([data[data['Country Name'] == 'Aruba']['Value'],
             data[data['Country Name'] == 'Zimbabwe']['Value']],
            labels=['Aruba', 'Zimbabwe'])
plt.xlabel('Country')
plt.ylabel('Value')
plt.title('Box Plot of Value by Country')
plt.show()
```

```python
    #Line Plot

fig, ax = plt.subplots()
for key, grp in data.groupby('Country Name'):
    ax.plot(grp['Year'], grp['Value'], label=key)
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Value Over Time by Country')
plt.legend()
plt.show()
```

```python
   #Displot

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
sns.displot(data, x='Value', kde=True)
plt.title('Distribution of Population Values')
plt.xlabel('Population Value')
plt.ylabel('Density')
plt.show()
```
