```python
import pandas as pd
df=pd.read_csv('/content/StudentsPerformance (1).csv')
print(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt

print(df.info())
print(df.isna().sum())
print(df.describe())

df=pd.read_csv('/content/StudentsPerformance (1).csv')

x=df[['reading score', 'writing score']]
y=df['math score']

x_train, x_test, y_train, y_test = train_test_split (x,y, test_size=0.2, random_state=42)

model=LogisticRegression()
lr=model.fit(x_train, y_train)
y_pred_lr=lr.predict(x_test)

print('mean_squared_error:',mean_squared_error(y_test,y_pred_lr))
print('accuracy_score:',accuracy_score(y_test,y_pred_lr))
print('confusion_matrix:',confusion_matrix(y_test,y_pred_lr))

plt.figure(figsize=(8,6))
plt.scatter(y_test,y_pred_lr, color='red')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC, SVR
from sklearn.metrics import mean_squared_error, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt

df=pd.read_csv('/content/StudentsPerformance (1).csv')

x=df[['reading score', 'writing score']]
y=df['math score']

x_train, x_test, y_train, y_test = train_test_split (x,y, test_size=0.2, random_state=42)

model=SVC()
svc=model.fit(x_train, y_train)
y_pred_svc=svc.predict(x_test)

print('mean_squared_error:',mean_squared_error(y_test,y_pred_svc))
print('accuracy_score:',accuracy_score(y_test,y_pred_svc))
print('confusion_matrix:',confusion_matrix(y_test,y_pred_svc))

plt.figure(figsize=(8,6))
plt.scatter(y_test,y_pred_svc, color='green')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

df=pd.read_csv('/content/StudentsPerformance (1).csv')

x=df[['reading score', 'writing score']]

wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters = i)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
print('wcss:',wcss)

plt.plot(range(1,11),wcss,color='orange', marker='*')
```
