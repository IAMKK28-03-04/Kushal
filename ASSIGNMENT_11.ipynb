```python
import nltk
import pandas as pd

df = pd.read_csv('/content/smsa_test_orig_backtranslated_en.csv')

documents = []
for index, row in df.iterrows():
    document = ""
    for column, value in row.items():
        document += f"{column}: {value}\n"
    documents.append(document)

for document in documents:
    print(document)
    print("\n")
```

```python
from nltk.tokenize import sent_tokenize
nltk.download('punkt')

tokenized_documents = []
for document in documents:
    tokens = sent_tokenize(document)
    tokenized_documents.append(tokens)

for tokens in tokenized_documents:
    print(tokens)
    print("\n")
```

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')

tokenized_documents = []
for document in documents:
    tokens = word_tokenize(document)
    tokenized_documents.append(tokens)

for tokens in tokenized_documents:
    print(tokens)
    print("\n")
```

```python
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

all_tokens = [token for document_tokens in tokenized_documents for token in document_tokens]

fdist = FreqDist(all_tokens)

print(fdist.most_common(10))

fdist.plot(30,cumulative=False)
plt.show()
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

filtered_tokens = []
for tokens in tokenized_documents:
    filtered_document_tokens = [token for token in tokens if token.lower() not in stop_words]
    filtered_tokens.append(filtered_document_tokens)

for tokens in filtered_tokens:
    print(tokens)
    print("\n")
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

filtered_tokens_spacy = []
for document in documents:
    doc = nlp(document)
    filtered_document_tokens = [token.text for token in doc if not token.is_stop]
    filtered_tokens_spacy.append(filtered_document_tokens)

for tokens in filtered_tokens_spacy:
    print(tokens)
    print("\n")
```

```python
import string

def remove_punctuation(text):
  """Removes punctuation from a string.

  Args:
    text: The string to remove punctuation from.

  Returns:
    The string with punctuation removed.
  """
  return text.translate(str.maketrans('', '', string.punctuation))

df['text_no_punct'] = df['text'].apply(remove_punctuation)

print(df)
```

```python
df['text_uppercase'] = df['text'].str.upper()

df['text_lowercase'] = df['text'].str.lower()

print(df)
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_tokens = []
lemmatized_tokens = []

for tokens in filtered_tokens_spacy:
    stemmed_document_tokens = [stemmer.stem(token) for token in tokens]
    stemmed_tokens.append(stemmed_document_tokens)

    lemmatized_document_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    lemmatized_tokens.append(lemmatized_document_tokens)


print("Stemmed tokens:")
for tokens in stemmed_tokens:
    print(tokens)
    print("\n")

print("Lemmatized tokens:")
for tokens in lemmatized_tokens:
    print(tokens)
    print("\n")
```

```python
import pandas as pd
from nltk import ngrams

def generate_ngrams(text, n):
  """Generates n-grams from a given text.

  Args:
    text: The input text.
    n: The size of the n-grams.

  Returns:
    A list of n-grams.
  """
  tokens = nltk.word_tokenize(text)
  return list(ngrams(tokens, n))


for feedback in df['text']:
  print(f"text: {text}")
  unigrams = generate_ngrams(feedback, 1)
  bigrams = generate_ngrams(feedback, 2)
  trigrams = generate_ngrams(feedback, 3)

  print(f"Unigrams: {unigrams}")
  print(f"Bigrams: {bigrams}")
  print(f"Trigrams: {trigrams}")
  print("\n")
```
